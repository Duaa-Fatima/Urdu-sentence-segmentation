SENTENCE SEGMENTATION

Word Splitting: The text is split into individual words, and empty spaces are removed.
Sentence Ender Detection: Checks for sentence-ending words or punctuation and adds a period (۔) if needed.
Rejoin Text: Words are rejoined into a single string, then split into sentences using periods.
Merge Short Sentences: Sentences shorter than 50 characters are merged with the previous sentence.
Clean Whitespace: Extra spaces are removed from the final list of sentences.
Save Output: The processed sentences are saved to a CSV file.

```python
def my_sentence_tokenizer(words):
   # Split the input into words
    words = str(words).split(" ")
    words = [x for x in words if x.strip()]  # Remove empty strings

    # Process words to add sentence finishers
    for index, x in enumerate(words):
        if index + 1 < len(words):
            if '؟' in str(x) or '!' in str(x):
                words.insert(index + 1, "۔")
            if x in sentence_enders:
                if words[index + 1] not in urdu_conjunctions:
                    words.insert(index + 1, "۔")

    # Join words back into a single string and split into sentences
    data = " ".join(words)
    data_lst = data.split("۔")

    # Combine short sentences into previous ones
    index = 1
    while index < len(data_lst):
        if len(data_lst[index].strip()) < 50:  # Check length of the sentence
            data_lst[index - 1] += data_lst[index]  # Merge with the previous sentence
            data_lst.pop(index)  # Remove the merged sentence
        else:
            index += 1  # Only increment if no merging

    # Clean up any extra spaces
    data_lst = [re.sub(r'\s+', ' ', x).strip() for x in data_lst]

    # Save sentences to a DataFrame and CSV
    df = pd.DataFrame({'Sentences': data_lst})
    df.to_csv("output.csv", index=False)

    return data_lst
```

```python
def calculate_accuracy(urduhack_segments, your_segments):
    # Convert both lists to sets for easier comparison
    urduhack_set = set(urduhack_segments)
    your_set = set(your_segments)

    # Calculate matches
    matches = urduhack_set.intersection(your_set)

    # Calculate accuracy
    accuracy = len(matches) / len(urduhack_set) * 100 if urduhack_set else 0
    return accuracy, matches
```

SENTENCE PIECE

 Build a vocabulary of subwords from the given Urdu sentences using a BPE-like approach.

```python
import json
```

```python
class SimpleSentencePieceWordWise:
    def __init__(self, vocab_file='vocab.json'):
        self.vocab = self.load_vocab(vocab_file)
        self.reverse_vocab = {v: k for k, v in self.vocab.items()}

    def load_vocab(self, vocab_file):
        with open(vocab_file, 'r', encoding='utf-8') as f:
            vocab = json.load(f)
        return vocab

    def tokenize(self, text):
        # Urdu-specific tokenization
        tokens = re.findall(r'\w+|[^\w\s]', text)
        return tokens

    def encode(self, sentence):
        tokens = self.tokenize(sentence)
        encoded = []
        i = 0

        while i < len(tokens):
            if i < len(tokens) - 1:
                bigram = f"{tokens[i]} {tokens[i+1]}"
                if bigram in self.vocab:
                    encoded.append(self.vocab[bigram])
                    i += 2
                    continue

            if tokens[i] in self.vocab:
                encoded.append(self.vocab[tokens[i]])
            else:
                encoded.append(self.vocab.get('<unk>', -1))
            i += 1

        return encoded

    def decode(self, encoded):
        decoded = []
        i = 0
        while i < len(encoded):
            if encoded[i] in self.reverse_vocab:
                token = self.reverse_vocab[encoded[i]]
                if ' ' in token:  # It's a bigram
                    decoded.extend(token.split())
                else:
                    decoded.append(token)
            else:
                decoded.append('<unk>')
            i += 1
        return ' '.join(decoded)

def build_and_save_vocab(sentences, vocab_size=5000, vocab_file='vocab.json'):
    vocab = {'<unk>': 0, '<s>': 1, '</s>': 2}
    tokens = []

    def tokenize(text):
        return re.findall(r'\w+|[^\w\s]', text)

    # Tokenize sentences and gather all tokens
    for sentence in sentences:
        token_list = tokenize(sentence)
        tokens.extend(token_list)

    # Count frequencies of individual words
    word_counts = Counter(tokens)

    # Add words to vocabulary
    for word, count in word_counts.most_common(vocab_size - len(vocab)):
        if word not in vocab:
            vocab[word] = len(vocab)

    # Now handle bigrams
    bigrams = [f"{tokens[i]} {tokens[i + 1]}" for sentence in sentences for tokens in [tokenize(sentence)] for i in range(len(tokens) - 1)]
    bigram_counts = Counter(bigrams)

    # Add bigrams to vocabulary
    for bigram, count in bigram_counts.most_common(vocab_size):
        if len(vocab) < vocab_size and bigram not in vocab:
            vocab[bigram] = len(vocab)

    # Save vocabulary to a file
    with open(vocab_file, 'w', encoding='utf-8') as f:
        json.dump(vocab, f, ensure_ascii=False, indent=2)

    print(f"Vocabulary saved: {vocab_file}")
    return vocab
```
